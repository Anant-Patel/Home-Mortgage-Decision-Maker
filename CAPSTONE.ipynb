```python
# Created function to look : datatype, percent missing, No.missing and No.Unique of each column to understand

def clean(df):
    with pd.option_context('display.max_rows', None, 'display.max_columns', None):
        info = pd.DataFrame()
        info['data type'] = df.dtypes    # Datatype of each column
        info['percent missing'] = df.isnull().sum()*100/len(df)   # Percentage of missing values of each column
        info['No. missing'] = df.isnull().sum()   # Total number of missing values of each column 
        info['No. unique'] = df.apply(lambda x: len(x.unique()))  # Total number of unique values of each column
        return info.sort_values('data type')  # Sorting output by datatype
```

```python
# Created function : Missing Values Handling

# Converting values that are equivalent to null or null(but considered as NON-NULL) into NAN values. 

from pandas.api.types import is_string_dtype, is_object_dtype

def nan(df):
    for col in df.columns:
        if is_string_dtype(df[col]) or is_object_dtype(df[col]):
            df[col] = df[col].str.lower()
            df[col] = df[col].fillna(np.nan)
            df[col] = df[col].replace('not applicable', np.nan)
            df[col] = df[col].replace('information not provided by applicant in mail, internet, or telephone application', np.nan)
            df[col] = df[col].replace('no co-applicant', np.nan)
#             df[col] = df[col].replace('??', np.nan)
            df[col] = df[col].replace('', np.nan)
```

```python
# One of the way to handle NAN values : Random Imputation

# Some string values containing features have NAN values
# If we drop out them, we could lost important information and there is at least one NAN in each and every row.
# So, we thought going by randomly handling would be good.
# For Ex: Male/Feamle : We neither do mean/median nor drop them.
import random

def random_nan(columns):
    for col in columns:
        df_list=home[col].unique().tolist()
        cleanedList = [x for x in df_list if str(x) != 'nan']
        home[col] = home[col].apply(lambda v: random.choice(cleanedList))
        
```

```python
# Converting all the string/categorical dtype column values into numeric

from pandas.api.types import is_categorical_dtype, is_string_dtype

def df_string_to_cat(df):
    for col in df.columns:
        if is_string_dtype(df[col]):
            df[col] = df[col].astype('category').cat.as_ordered()
            
def df_cat_to_catcode(df):
    for col in df.columns:
        if is_categorical_dtype(df[col]):
            df[col] = df[col].cat.codes + 1
```

```python
# Feature Selection technique : Univariate selection 

import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2


x = home.drop(['action_taken_name'], axis=1)  #independent columns
y = home['action_taken_name']   #target column i.e price range


#apply SelectKBest class to extract top 10 best features
bestfeatures = SelectKBest(score_func=chi2, k=10)
fit = bestfeatures.fit(x,y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(x.columns)
#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  #naming the dataframe columns
print(featureScores.nlargest(10,'Score'))  #print 10 best features
```

```python
# Feature Selection technique : ExtraTreesClassifier

import pandas as pd
import numpy as np
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt


model = ExtraTreesClassifier()
model.fit(x,y)
print(model.feature_importances_) #use inbuilt class feature_importances of tree based classifiers
#plot graph of feature importances for better visualization
feat_importances = pd.Series(model.feature_importances_, index=x.columns)
feat_importances.nlargest(68).plot(kind='barh')
plt.show()
```

```python
x = home.drop(['action_taken_name'], axis=1)      # Feature Variables
y = home['action_taken_name']                     # Target Variable
```

```python
# splitting the dataset into training set and test set

from sklearn.model_selection import train_test_split # used for splitting training and testing data

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler # used for feature scaling

# feature scaling

sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
import pickle

with open('xtrain.pickle', 'wb') as xtn:
    pickle.dump(X_train, xtn)
with open('xtest.pickle', 'wb') as xts:
    pickle.dump(X_test, xts)
```

```python
import pickle 

with open('xtrain.pickle', 'rb') as xtn:
    X_train = pickle.load(xtn)
with open('xtest.pickle', 'rb') as xts:
    X_test = pickle.load(xts)
```

```python
import xgboost as xgb

# metrics
from sklearn.metrics import accuracy_score 

xg = xgb.XGBClassifier(n_estimators=100, learning_rate=0.08, gamma=0, subsample=0.75,
                           colsample_bytree=1, max_depth=7)
xg.fit(X_train, Y_train.squeeze().values)

y_train_pred1 = xg.predict(X_train)
y_test_pred1 = xg.predict(X_test)
```

```python
print('XGB accuracy score for train:',
        accuracy_score(Y_train, y_train_pred1),
        accuracy_score(Y_test, y_test_pred1))
```

```python
xgtrain = xg.score(X_train, Y_train)
xgtest = xg.score(X_test, Y_test)
```

```python
# importing necessary libraries and functions
import numpy as np
from flask import Flask, request, jsonify, render_template
import pickle

app = Flask(__name__, template_folder='template') #Initialize the flask App
model = pickle.load(open('model.pkl', 'rb')) # loading the trained model

@app.route('/') # Homepage
def home():
    return render_template('index.html')

@app.route('/predict',methods=['POST'])
def predict():
    '''
    For rendering results on HTML GUI
    '''
    
    # retrieving values from form
    init_features = [float(x) for x in request.form.values()]
    final_features = [np.array(init_features)]

    prediction = model.predict(final_features) # making prediction
    return render_template('./index.html', prediction_text='Predicted Class: {}'.format(prediction)) # rendering the predicted result

if __name__ == "__main__":
    app.run(debug=False)
```
